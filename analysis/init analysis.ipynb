```python
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Ignore font warnings
warnings.filterwarnings('ignore')

# Use standard English fonts
plt.rcParams['font.sans-serif'] = ['Arial', 'Helvetica', 'sans-serif']

# Create a 1x2 figure (width 14, height 6)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# ==========================================
# Chart 1: Null Field Ratio
# ==========================================
tables = ['ChintaroClient\n(99.4%)', 'Epicor_Clients\n(98.1%)', 'GoldCareClients\n(100%)', 'Leecare Residents\n(5.6%)']
null_rates = [99.4, 98.1, 100.0, 5.6]
colors_bar = ['#4C72B0', '#55A868', '#C44E52', '#8172B2']

bars = axes[0].bar(tables, null_rates, color=colors_bar, edgecolor='black', alpha=0.8)
axes[0].set_title('Null Field Proportion by Table', fontsize=15, fontweight='bold', pad=15)
axes[0].set_ylabel('Proportion of Null Fields (%)', fontsize=12)
axes[0].set_ylim(0, 115) # Leave space for text labels on top

# Add percentage labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2, yval + 1.5, f'{yval}%', 
                 ha='center', va='bottom', fontsize=12, fontweight='bold')

# ==========================================
# Chart 2: Epicor vs GoldCare Overlap (Pie Chart)
# ==========================================
overlap_count = 1800
only_epicor_count = 2000 - 1800    # 200
only_goldcare_count = 10000 - 1800 # 8200

labels = [f'Epicor Only\n({only_epicor_count})', 
          f'GoldCare Only\n({only_goldcare_count})', 
          f'Overlap (Duplicated)\n({overlap_count})']
sizes = [only_epicor_count, only_goldcare_count, overlap_count]
colors_pie = ['#6baed6', '#74c476', '#fd8d3c']
explode = (0, 0, 0.1)  # Slightly separate the "Overlap" slice to highlight it

axes[1].pie(sizes, explode=explode, labels=labels, colors=colors_pie, autopct='%1.1f%%',
        shadow=True, startangle=140, textprops={'fontsize': 13})
axes[1].set_title('Epicor vs GoldCare Customer Overlap', fontsize=15, fontweight='bold', pad=15)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
